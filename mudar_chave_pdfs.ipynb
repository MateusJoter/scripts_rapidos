```python
import fitz  # PyMuPDF
import os
```

```python
def remover_chave_em_pdf(caminho_pasta, chave):
    # Percorre todas as subpastas
    for raiz, pastas, arquivos in os.walk(caminho_pasta):
        for nome_arquivo in arquivos:
            if nome_arquivo.lower().endswith(".pdf"):
                caminho_completo = os.path.join(raiz, nome_arquivo)
                
                try:
                    doc = fitz.open(caminho_completo)
                    alterado = False
                    
                    for pagina in doc:
                        # Busca todas as instâncias da chave
                        lista_instancias = pagina.search_for(chave)
                        
                        if lista_instancias:
                            alterado = True
                            # "Apaga" desenhando um retângulo branco sobre o texto
                            for inst in lista_instancias:
                                pagina.add_redact_annot(inst, fill=(1, 1, 1))
                            pagina.apply_redactions()
                    
                    if alterado:
                        # Salva o arquivo modificado (sobrescreve o original)
                        doc.save(caminho_completo, incremental=True, encryption=fitz.PDF_ENCRYPT_KEEP)
                        print(f"Sucesso: '{chave}' removida de {nome_arquivo}")
                    
                    doc.close()
                except Exception as e:
                    print(f"Erro ao processar {nome_arquivo}: {e}")
```

```python
# Configuração
diretorio_alvo = r""  # Coloque aqui o caminho da sua pasta raiz
chave_para_remover = ""

remover_chave_em_pdf(diretorio_alvo, chave_para_remover)
```
